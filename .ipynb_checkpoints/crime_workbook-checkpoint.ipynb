The following project is the code for submission to the following kaggle competition: https://www.kaggle.com/c/sf-crime

The synopsis of the problem is below:

Predict the category of crimes that occured in SF based on the features below
Features:
Dates - timestamp of the crime incident
Category - category of the crime incident (only in train.csv). This is the target variable you are going to predict.
Descript - detailed description of the crime incident (only in train.csv)
DayOfWeek - the day of the week
PdDistrict - name of the Police Department District
Resolution - how the crime incident was resolved (only in train.csv)
Address - the approximate street address of the crime incident 
X - Longitude
Y - Latitude

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loc="C:\Users\user\Desktop\kagglehelp\crime"
loc=loc.replace("\\", "\\\\")
print loc
```

```python
data=pd.read_csv(loc+'\\train.csv')
data.describe()
```

```python
data.head()
```

```python
data.dtypes
```

```python
# convert strings into labelcodes
from sklearn.preprocessing import LabelEncoder

cat=LabelEncoder()
desc=LabelEncoder()
day=LabelEncoder()
dist=LabelEncoder()
res=LabelEncoder()
data_cat=cat.fit_transform(data[data.columns[1]])
data_le=pd.DataFrame(data=data_cat, columns=["Category"])


data_desc=desc.fit_transform(data[data.columns[2]])
data_le["Description"]=data_desc
data_day=day.fit_transform(data[data.columns[3]])
data_le["DayOfWeek"]=data_day
data_dist=dist.fit_transform(data[data.columns[4]])
data_le["PdDistrict"]=data_dist
data_res=res.fit_transform(data[data.columns[5]])
data_le["Resolution"]=data_res

data_le

# data_le=data[data.columns[1:-3]].apply(le.fit_transform)
data_nle=data[['Dates', 'Address', 'X', 'Y']]
data_adj=pd.concat([data_le, data_nle], axis=1)
```

```python
# Find the Month and hours from from Dates and set them as a seperate variable

data_adj['Hour']=pd.to_datetime(data.Dates).dt.hour
data_adj["Month"]=pd.to_datetime(data.Dates).dt.month
```

```python
# make exploratory plots
%matplotlib inline
data_adj.hist(figsize=(20,20))
```

```python
data_adj.columns
```

```python
# Lets implement cross validation for test train split
from sklearn.cross_validation import train_test_split
# lets drop Dates
data_adj2=data_adj.drop('Dates', 1)
data_adj2=data_adj2.drop('Address', 1)
data_adj2=data_adj2.drop('Description', 1)
data_adj2=data_adj2.drop('Resolution', 1)
y=data_adj2.values[:,0]
X=data_adj2.values[:,1:]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
data_adj2.columns
```

```python
# Let us now try to fit the naive bayes model using newly created training sets
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# Let us now try to fit the RandomForest Classifier model

from sklearn.ensemble import RandomForestClassifier
clf2=RandomForestClassifier(15)
clf2.fit(X_train, y_train)
clf2.score(X_test, y_test)
```

```python
# Random Forests works better
clf2.feature_importances_
# it looks like the location(x,y) hour and month are very helpful in prediction
```

```python
p=clf2.predict_proba(X_test)
```

```python
# Lets setup for submission
test=pd.read_csv(loc+'//test.csv')

test["Hour"]=pd.to_datetime(test.Dates).dt.hour
test["Month"]=pd.to_datetime(test.Dates).dt.month
test_adj=test.drop('Dates', 1)
test_adj=test_adj.drop('Address', 1)
```

```python
test_adj.head()
```
